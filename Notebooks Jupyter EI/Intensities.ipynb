```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
raw_0 = np.rot90(np.genfromtxt("1. RAW/RAW_sangre.txt"),1)
mean_raw = np.mean(raw_0)
for i in range(256):
    for j in range(256):
        if raw_0[i,j] == 0:
            raw_0[i,j] = mean_raw

plt.imshow(raw_0, cmap="bone")
```

```python
ff_0 = np.rot90(np.genfromtxt("2. FF/FF_sangreP6.txt"),1)
mean_ff = np.mean(ff_0)
for i in range(256):
    for j in range(256):
        if ff_0[i,j] == 0:
            ff_0[i,j] = mean_ff

plt.imshow(ff_0, cmap="bone")
```

```python
img_0 = ff_0/raw_0
plt.imshow(img_0, cmap="bone")
```

```python
even = img_0[:,::2]
odd = img_0[:,1::2]

phase = (even-odd)/(even+odd)
plt.figure(figsize=(10,10))
im = plt.imshow(phase, cmap="bone")
plt.xlabel("Nmero de pxel en el eje x del detector")
plt.ylabel("Nmero de pxel en el eje y del detector")
plt.title('Imagen de fase por Navarrete para la muestra\n de sangre artificial del tubo de 1.54 mm\n', fontsize=14)
plt.gca().invert_yaxis()
clb = plt.colorbar(im,fraction=0.046, pad=0.04)
clb.ax.get_yaxis().labelpad = 15
clb.set_label('Intensidad [a.u.]', rotation=270, fontsize=12)
plt.tight_layout()
plt.savefig("peq.pdf")
```

```python
perfiles_peq = phase[0,:]
for i in range(1,256): 
    perfiles_peq += phase[i,:]
```

```python
plt.figure(figsize=(12,8))
plt.plot(np.abs(perfiles_peq), c="green")
plt.title("Perfil de intensidad promedio por Navarrete para el tubo de 1.43 mm relleno de sangre artificial")
plt.ylabel("Intensidad promedio")
plt.xlabel("Nmero de pxel en el detector")
```

```python
I_max = max(abs(phase).max(axis=1))
I_bkg1 = max(abs(phase[:, 0:60]).max(axis=1))
sigma_bkg1 = (abs(phase[:, 0:60]).max(axis=1)).std()
I_bkg2 = max(abs(phase[:, 90:]).max(axis=1))
sigma_bkg2 = (abs(phase[:, 90:]).max(axis=1)).std()
I_bkg = (I_bkg1 + I_bkg2)/2
sigma_bkg = (sigma_bkg1 + sigma_bkg2)/2

CNR_sangre_1 = (I_max - I_bkg)/sigma_bkg
CNR_sangre_1
```
